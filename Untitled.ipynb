```python
# Get 2024 season schedule for Washington Nationals (team ID: 120)
nationals_schedule = statsapi.schedule(
    team=120, 
    start_date='2024-03-28', 
    end_date='2024-11-02'
)

# Print each game with more detailed information
for game in nationals_schedule:
    home_team = game['home_name']
    away_team = game['away_name']
    game_date = game['game_date']
    game_time = game.get('game_time', 'TBD')
    venue = game['venue_name']
    
    # Format the output to show if Nationals are home or away
    if home_team == "Washington Nationals":
        print(f"{game_date} - {away_team} @ Washington {game_time} ({venue})")
    else:
        print(f"{game_date} - Washington @ {home_team} {game_time} ({venue})")
```

```python
game_id = 744798

# 1. Get the line score
print("LINESCORE:")
print("-" * 50)
print(statsapi.linescore(game_id))
print("\n")

# 2. Get the box score
print("BOXSCORE:")
print("-" * 50)
print(statsapi.boxscore(game_id))
print("\n")

# 3. Get scoring plays
print("SCORING PLAYS:")
print("-" * 50)
print(statsapi.game_scoring_plays(game_id))
print("\n")

# 4. Get highlights
print("HIGHLIGHTS:")
print("-" * 50)
print(statsapi.game_highlights(game_id))
```

```python
# Get raw boxscore data
boxscore_data = statsapi.boxscore_data(744798)

# Let's look at the structure of the data
import json
print(json.dumps(boxscore_data, indent=2))
```
